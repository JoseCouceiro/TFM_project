```python
import os
import json
import numpy as np
import pandas as pd
```

```python
drugs = pd.read_csv('pubchem_dataset_label_clean.csv')
```

```python
# Set lipinski rules
hdonor = drugs['HBondDonorCount'] < 6
haccept = drugs['HBondAcceptorCount'] < 10
mw = drugs['MolecularWeight'] < 500
clogP = drugs['LogP'] < 5
# Apply rules to dataframe
drugs['RuleFive'] = np.where(((hdonor & haccept & mw) | (hdonor & haccept & clogP) | (hdonor & mw & clogP) | (haccept & mw & clogP)), 1, 0)
```

```python
with open(os.path.join('..','dataframes_resources','dataframes_props.json')) as file:
    res = json.load(file)
column_order = res['column_order']
```

```python
drugs = drugs[column_order]
drugs.head()
```

```python
drugs['ATC_Code'] = drugs['ATC_Code'].str[0]
```

```python
drugs.to_csv(os.path.join('pubchem_dataset_label_clean.csv'), index= False)
```
