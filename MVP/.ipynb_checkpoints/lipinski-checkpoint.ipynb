## Analysis of the predictions and their relation to Lipinski rule

Lipinski's rule of 5 is an evaluation of the druglikeness of molecular compounds based on the empirical observation of the characteristics of orally administered drugs. The molecular properties described in this rule have important implications in the pharmacokinetics of drugs, which includes absorption, distribution, metabolism, and excretion ("ADME"). The rule states that drugs have no more than one violation of these criteria:
- No more than 5 hydrogen bond donors
- No more than 10 hydrogen bond acceptors
- Molecular mass of less than 500 daltons
- A calculated octanol-water partition coefficient (Clog P) not exceeding 5

Next, I'm going to analyse if our predictions have been biased in some way among drugs that comply Lipinski's rule and drugs that do not.

### Imports

```python
import os
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
# %matplotlib inline
```

### Load Data

Load raw dataframe

```python
drugs = pd.read_csv(os.path.join('res','raw_data', 'CID_properties_nr.csv'))
# Select Lipinski features
lipinski = drugs.drop(['IsomericSMILES', 'desalted_SMILES'], axis=1)
```

Obtain the CID for the predictions

```python
# Train/test split with the same seed used for training our model
X_train, X_test, y_train, y_test = train_test_split(drugs, drugs['drug_class'], test_size=0.3, random_state=42)
cids = X_test.iloc[1000:]['CID'].reset_index().drop('index', axis=1)
```

```python
# Load the predictions
y_pred = pd.read_pickle(os.path.join('predictions', 'y_preds.pkl'))
y_pred.name = 'predicted_class_code'
```

```python
# Concat with CID
cid_pred = pd.concat((cids, y_pred), axis=1)
```

```python
# Merge with lipinski dataframe
lipinski_pred = lipinski.merge(cid_pred, right_on='CID', left_on='CID')
lipinski_pred.sort_values('CID').head()
```

### Identify bad predictions

Use the dictionary of class codes to make a 'predicted class' column and, by comparison with 'drug class', obtain a column that indicates if the prediction was correct or not

```python
with open(os.path.join('res','config', 'class_codes.json')) as file:
    codes_dic = json.load(file)
inv_codes_dic = {value: key for key, value in codes_dic.items()}
```

```python
lipinski_pred['predicted_class'] = lipinski_pred['predicted_class_code'].map(lambda x: codes_dic[str(int(x))])
lipinski_pred['bad_preds'] = lipinski_pred['predicted_class']==lipinski_pred['drug_class']
lipinski_pred
```

### Calculate the rate of error among lipinki compliers and non compliers

```python
# Set lipinski rules
hdonor = lipinski_pred['HBondDonorCount'] < 6
haccept = lipinski_pred['HBondAcceptorCount'] < 10
mw = lipinski_pred['MolecularWeight'] < 500
clogP = lipinski_pred['XLogP'] < 5
# Apply rules to dataframe
lipinski_pred_compliers = lipinski_pred[(hdonor & haccept & mw) | (hdonor & haccept & clogP) | (hdonor & mw & clogP) | (haccept & mw & clogP)]

# Count and calculate ratios
n_correct_compliers = lipinski_pred_compliers['bad_preds'].sum()
n_correct_total = lipinski_pred['bad_preds'].sum()

accuracy_compliers = n_correct_compliers/len(lipinski_pred_compliers)
accuracy_total = n_correct_total/len(lipinski_pred)

print(f'Error rate among Lipinski compliers: {(1-accuracy_compliers):.3f}')
print(f'Total error rate: {(1-accuracy_total):.3f}')
```

It seems that the model is slightly more prone to innacurately predict drugs classes among Lipinski compliers. Considering this result and that the properties used by Lipinski's rule are codified in the molecular fingerprints, we may assume that these properties are not a principal consideration for our prediction model, and other molecular characteristics seem to have more weight. This is expected since all the molecules analysed ara actually drugs, if the model were trained including molecules that are not drugs, the Lipinski properties probably would have an important weight.

Below there is a graphical representation of prediction errors among Lipinski compliers and non compliers:

```python
fig, axs = plt.subplots(1,2, figsize=(12,6))

# H Bond acceptor count v H Bond Donor count
sns.scatterplot(ax= axs[0], data=lipinski_pred, x='XLogP', y='HBondAcceptorCount',hue='bad_preds')
axs[0].axvline(x = 5, ymin = 0, ymax=10)
axs[0].axhline(y = 10, xmin = 0, xmax=50)
axs[0].legend('', frameon=False)
axs[0].set_ylabel('Clog P')
axs[0].set_xlabel('H Bond Acceptor Count')
axs[0].annotate('Non Compliers', (10,10.5), color = 'blue', fontsize = 10)

#plt.setp(axs[0].xaxis.get_majorticklabels(), rotation=90, ha='left')

# Molecular weight v Clog P
sns.scatterplot(ax= axs[1], data=lipinski_pred, x='MolecularWeight', y='HBondDonorCount',hue='bad_preds')
axs[1].axvline(x = 500, ymin = 0, ymax=lipinski_pred['XLogP'].max())
axs[1].axhline(y = 5, xmin = 0, xmax=50)
axs[1].set_ylabel('H Bond Donor Count')
axs[1].set_xlabel('Molecular Weight')
axs[1].annotate('Non Compliers', (2000,5.5), color = 'blue', fontsize = 10)

# Legend
labels = ['Mistake', 'Correct']
c, l = axs[1].get_legend_handles_labels()
axs[1].legend(c, labels, loc='upper right')

plt.tight_layout()
plt.show()
```
