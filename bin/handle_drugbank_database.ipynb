```python
with open(os.path.join('drugbank_all_full_database.xml', 'full database.xml')) as drugs_file:
    lines = drugs_file.readlines()
```

```python
two_instances = lines[0:8234]
two_i = open(os.path.join('drugbank_all_full_database.xml', 'two_instances.xml'), 'w')
for line in two_instances:
    two_i.write(line)
two_i.write('</drug>\n \n')
two_i.close()
```

```python
two_last_instances = lines[-8234:-1]
two_last_i = open(os.path.join('drugbank_all_full_database.xml', 'two_last_instances.xml'), 'w')
for line in two_last_instances:
    two_last_i.write(line)
two_last_i.close()
```
