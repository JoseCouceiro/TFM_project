```python
toy_drugs = drugs
toy_drugs['compound'] = toy_drugs['cid'].map(pcp.Compound.from_cid)
toy_drugs.head()
```

```python
toy_drugs[[
    'HBondAcceptorCount',
    'HBondDonorCount',
    'MolecularWeight',
    'LogP',
    'IsomericSMILES']] = toy_drugs['compound'].apply(lambda x: pd.Series([
        x.h_bond_acceptor_count,
        x.h_bond_donor_count, 
        x.molecular_weight,
        x.xlogp,
        x.isomeric_smiles]))
toy_drugs
```
